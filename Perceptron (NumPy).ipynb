```python
inputs_vec = np.array([1,2,3])
weights_vec = np.array([0.5, 0.8, 0.2])
bias = 0.81

summ = (inputs_vec @ weights_vec.T) + bias
print(summ)
```

#### _Activation Function_

* Weighted summation gives the non-linearity to the neural network which enables the model to learn complex data patterns.

* Without the Activation Function the model will behave like simple linear regression model.

* Activation functions decide whether a neuron should be activated based on the weighted sum of inputs and a bias term.
