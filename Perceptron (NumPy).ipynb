#### Weighted summation

```python
inputs_vec = np.array([1,2,3])
weights_vec = np.array([0.5, 0.8, 0.2])
bias = 0.81
def multi():
    summation = 0
    for x,w in zip(inputs_vec, weights_vec):
        summation += x*w
    return summation

weighted_summation = multi() + bias
print(weighted_summation)
```
